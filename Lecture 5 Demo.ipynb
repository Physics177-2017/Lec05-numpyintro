```python
x = np.linspace(0,2,100)
# array of sample points

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quad')
plt.plot(x, x**3, label='cube')

plt.xlabel('x label')
plt.ylabel('function value')
plt.title("Plot!")
plt.legend()
plt.show()
```

# Vectorize

```python
from math import sin as sin
def mySine(x):
    return sin(x)

mySine(3.14/2)
```

```python
xshort = np.linspace(0,2,4)
plt.plot(xshort, mySine(xshort))
plt.show()
```

```python
myNumpySine = np.vectorize(mySine)
xshort = np.linspace(0,3,7)

plt.plot(xshort, myNumpySine(xshort))
plt.show()
```

```python
def weirdSine(k, x):
    return sin(k*x)
    
myWeirdSine = np.vectorize(weirdSine)

xshort = np.linspace(0,6,100)
plt.plot(xshort, myWeirdSine(1,xshort))
plt.plot(xshort, myWeirdSine(2,xshort))
plt.plot(xshort, myWeirdSine(3,xshort))
plt.plot(xshort, myWeirdSine(4,xshort))
plt.show()
```
