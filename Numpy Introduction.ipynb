## Vectorize
This is where the real power of the arrays show up.  

What if we have a more complicated function?

```python
from math import sin as sin
def mySine(x):
    return sin(x)

mySine(3.14/2)
```

```python
xshort = np.linspace(0, 2, 4)
plt.plot(xshort, mySine(xshort))
# Doesn't work, because mySine doesn't know how to take numpy arrays!
plt.show()
```

```python
# create a vectorized function that takes in numpy arrays
myNumpySine = np.vectorize(mySine)
xshort = np.linspace(0, 3, 7)

plt.plot(xshort, myNumpySine(xshort))
plt.title("Vectorized function")
plt.show()
```

```python
# This also works for multipl arguments
def weirdSine(k,x):
    return sin(k*x)


myWeirdSine = np.vectorize(weirdSine)

x = np.linspace(0,6,10)
x = np.linspace(0,6,100)

plt.plot(x, myWeirdSine(1,x), label='$k=1$')
plt.plot(x, myWeirdSine(2,x), label='$k=3$')
plt.plot(x, myWeirdSine(3,x), label='$k=2$')
plt.title("Vectorized function")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
plt.show()
```
